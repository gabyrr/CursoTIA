## Explorando datasets
Notebook perteneciente al curso: Tópicos de inteligencia artificial: de métodos clásicos a modelos generativos. Por Gabriela Ramírez de la Rosa

```python
#Explorando un dataset

from sklearn.datasets import load_iris
iris_dataset = load_iris()

X = iris_dataset.data
y = iris_dataset.target

## obteniendo los nombres de los atributos de cada instancia
print(iris_dataset.feature_names)
```

```python
## Cómo se ve el dataset en dos dimensiones?
import matplotlib.pyplot as plt

_, ax = plt.subplots()
scatter = ax.scatter(X[:, 0], X[:, 1], c=y)
ax.set(xlabel=iris_dataset.feature_names[0], ylabel=iris_dataset.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris_dataset.target_names, loc="lower right", title="Classes"
)
```

Una forma de ver todas las características en un conjunto de dimensiones reducido es utilizar herramientas como PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(X)

##por pasos, la instrucción anterior es igual a:
#pca = PCA(n_components=3)
#pca.fit(X)
#X_reduced = pca.transform(X)

ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])

plt.show()
```

Otras formas de visualizar el dataset Iris

```python
from mlxtend.plotting import scatterplotmatrix


#names = ['sepal length [cm]', 'sepal width [cm]',
#         'petal length [cm]', 'petal width [cm]']

names = iris_dataset.feature_names

fig, axes = scatterplotmatrix(X[y==0], figsize=(10, 8), alpha=0.5, names=names)
fig, axes = scatterplotmatrix(X[y==1], fig_axes=(fig, axes), alpha=0.5, names=names)
fig, axes = scatterplotmatrix(X[y==2], fig_axes=(fig, axes), alpha=0.5, names=names)

plt.tight_layout()
plt.show()
```
